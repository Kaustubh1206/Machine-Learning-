# Logistic Regression Implementation

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
# for cretating classification datasets
from sklearn.datasets import make_classification

```

```python
# Create Dataset 
X,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=8)
# X independent 
# y dependent 
# n_ smaple 1000 data points 
# classes -> 2 for binary classification 

```

```python
X
```

```python
pd.DataFrame(X) # TO see our independent features in table form
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train.shape
```

```python
# MODEL TRAINING 
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred=logistic.predict(X_test)
print(y_pred)
```

```python
# logistic.predict_proba(X_test) for probaility 
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report


```

```python
score=accuracy_score(y_test,y_pred)
print("accuracy_score",score)
cm=confusion_matrix(y_test,y_pred)
print("confusion Matrix: \n",cm)
print(classification_report(y_test,y_pred))
```

## Hyperparameter Tuning and Cross Validation 
 

## Grid Search CV - takes all combination 

```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg', 'lbfgs','liblinear', 'sag', 'saga'] 
```

```python
params=dict(penalty=penalty,C=c_values ,solver=solver)
```

```python
# GridSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold() # in this cv each fold has a balance propertiesb of each 
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)

# estimator = model 
# param_grid = params 
# n_jobs=-1 uses all cpu cores to spped up 
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)  
```

```python
score=accuracy_score(y_pred,y_test)
print("accuracy_score",score)
cm=confusion_matrix(y_pred,y_test)
print("confusion Matrix: \n",cm)
print(classification_report(y_pred,y_test))
```

## Randomised Search CV 
- takes Random attributes/parameters 

```python
from sklearn.model_selection import RandomizedSearchCV
model=LogisticRegression()
randomcv=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,scoring='accuracy')
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_pred=randomcv.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print("accuracy_score",score)
cm=confusion_matrix(y_pred,y_test)
print("confusion Matrix: \n",cm)
print(classification_report(y_pred,y_test))
```

## Logistic Regression for Multi Class Classifcation 

```python
X,y=make_classification(n_samples=1000,n_features=10,n_informative=3,n_classes=3,random_state=15)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression(multi_class='ovr')
logistic.fit(X_train,y_train)
y_pred=logistic.predict(X_test)
```

```python
y_pred
```

```python
score=accuracy_score(y_pred,y_test)
print("accuracy_score",score)
cm=confusion_matrix(y_pred,y_test)
print("confusion Matrix: \n",cm)
print(classification_report(y_pred,y_test))
```

## Hyperparameter Tunning 

```python
# Grid Search CV
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold


model=LogisticRegression()
param_grid =["l1","l2","elasticnet"]
c_values=[100,10,1,0.1,0.01]
solver=['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
params=dict(penalty=penalty,C=c_values ,solver=solver)

cv=StratifiedKFold()
grid=GridSearchCV(estimator=model,param_grid=params,scoring="accuracy",cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print("accuracy_score",score)
cm=confusion_matrix(y_pred,y_test)
print("confusion Matrix: \n",cm)
print(classification_report(y_pred,y_test))
```
