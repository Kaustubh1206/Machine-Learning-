```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Example scores (replace these with your actual results)
model_scores = {
    'Logistic Regression': {'Accuracy': 0.85, 'Precision': 0.82, 'Recall': 0.80, 'F1 Score': 0.81},
    'Random Forest': {'Accuracy': 0.90, 'Precision': 0.88, 'Recall': 0.87, 'F1 Score': 0.87},
    'Naive Bayes': {'Accuracy': 0.78, 'Precision': 0.76, 'Recall': 0.75, 'F1 Score': 0.75},
    'XGBoost': {'Accuracy': 0.92, 'Precision': 0.91, 'Recall': 0.89, 'F1 Score': 0.90}
}
```

```python
df_scores = pd.DataFrame(model_scores).T.reset_index()
df_scores = df_scores.rename(columns={'index': 'Model'})
```

```python
df_melted = df_scores.melt(id_vars='Model', var_name='Metric', value_name='Score')
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x='Model', y='Score', hue='Metric')
plt.title('Model Comparison - Accuracy, Precision, Recall, F1 Score')
plt.ylim(0, 1)
plt.legend(loc='lower right')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```
